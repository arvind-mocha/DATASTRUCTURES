# SINGLY LINKED LIST

#### 1. We need to create a node class
#### 2. The Create a Linkded List class which denotes the data inside the node
#### 3. Initially there is no node, Since there is no data
#### 4. create a head which acts as the first node and it is initially null since no data inside the any of the node in the linked list
#### 5. create a tail which acts as the last node and it is initially null since no data inside the any of the node in the linked list
#### 6. create a variable size which is initially 0 since there are no nodes inside the linked list
#### 7. create a method which return the length of the linked list
#### 8. create a method to find wheather the linked list is empty or not
#### 9. then create a method to insert element into our linked list
#### 10. elements are always inserted from the last
#### 11. create a function for displaying the linked list
#### 12. create a funtion for searching elements in the linked list
#### 13. if element found return the index of the element else point the the next node and increment the index by one
#### 14. Inserting an element as a first node, if isempty the head and tail points this new node
#### 15. if not isempty the next element is the head which is the current first node and the new node is the head which is inserted
#### 16. Inserting element at any index can be achieved by navigating our new node to the next of old node which is in that current position
#### 17. we must point the next of the old node to the new node and the next of new node to the node which is after the old node before insertion
#### 18. Deleting the first node can be achieved by just pointing the head node to the next node
#### 19. Deleting the last node can be achieved by traversing to the Node before the last node and setting it as the tail node and making tail Node's next to Null since there must be no node after tail Node
#### 20. Deleting Node at any position

```python
class _Node:
    __slots__ = '_element','_next'    # Memory Allocation
    def __init__(self,element,next):
        self._element = element
        self._next = next
        
class LinkedList:
    def __init__(self): # constructor
        self._head = None
        self._tail = None
        self._size = 0
        
    def __len__(self): # default super method which denotes the functionality of len function
        return self._size
    
    def isempty(self):
        return self._size == 0
    
    def addfirst(self,e):
        newest = _Node(e,None)
        if self.isempty():
            self._head = newest
            self._tail = newest
        else:
            newest._next = self._head
            self._head = newest
        self._size += 1
    
    def addlast(self,e):
        newest = _Node(e,None)
        if self.isempty():
            self._head = newest
        else:
            self._tail._next = newest
        self._tail = newest
        self._size += 1
        
    def addanywhere(self,e,position):
        newest = _Node(e,None)
        p = self._head
        i = 1
        while i < position:
            p = p._next
            i += 1
        newest._next = p._next
        p._next = newest
        self._size += 1
        
    
    def search(self, key):
        p = self._head
        index = 0
        
        while p:
            if p._element == key:
                return index
            p = p._next
            index += 1
        return 'element not found'
    
    def removefirst(self):
        if self.isempty():
            print('List is empty')
            return
        e = self._head._element
        self._head = self._head._next
        self._size -= 1
        if self.isempty():
            self._tail = None
        return e
    
    def removeanywhere(self,position):
        if self.isempty():
            print('List is empty')
            return
        p = self._head
        index = 1
        while index < position:
            p = p._next
            index += 1
        e = p._next._element
        p._next = p._next._next
        self._size -= 1
        
        return e
    
    def removelast(self):
        if self.isempty():
            print('List is empty')
            return
        p = self._head
        i = 1
        while i < self._size -1:
            p = p._next
            i += 1
        self._tail = p
        p = p._next
        e = p._element
        self._tail._next = None
        self._size -= 1
        return e
    
    def display(self):
        p = self._head
        while p:
            print(p._element,end=' --> ')
            p = p._next
        print()
        
    def insertsorted(self, e):
        newest = _Node(e, None)
        if self.isempty():
            self._head = newest
        else:
            p = self._head
            q = None
            while p and p._element < e:
                q = p
                p = p._next
            if p == self._head:
                newest._next = self._head
                self._head = newest
            else:
                newest._next = q._next
                q._next = newest
        self._size += 1
```

# CIRCULAR LINKED LIST

#### 1. Circular linked list is as same as single linked list. the only difference is the next of tail is the head node
#### 2. This refers ther is no start and end. But we need to keep a reference of head and tail just for traversing purpose
#### 3. if the circular linked list as only one node then the next of head is the head itself
#### 4. we need to have a counter for display method since there is no end and in singly linked list we dont need counter for display

```python
class _Node:
    __slots__ = '_element', '_next'

    def __init__(self, element, next):
        self._element = element
        self._next = next

class CircularLinkedList:
    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def isempty(self):
        return self._size == 0

    def addlast(self, e):
        newest = _Node(e, None)
        if self.isempty():
            newest._next = newest
            self._head = newest
        else:
            newest._next = self._tail._next
            self._tail._next = newest
        self._tail = newest
        self._size += 1

    def addfirst(self,e):
        newest = _Node(e, None)
        if self.isempty():
            newest._next = newest
            self._head = newest
            self._tail = newest
        else:
            self._tail._next = newest
            newest._next = self._head
            self._head = newest
        self._size += 1

    def addanywhere(self, e, position):
        newest = _Node(e, None)
        p = self._head
        i = 1
        while i < position - 1:
            p = p._next
            i = i + 1
        newest._next = p._next
        p._next = newest
        self._size += 1

    def removefirst(self):
        if self.isempty():
            print('List is empty')
            return
        e = self._head._element
        self._tail._next = self._head._next
        self._head = self._head._next
        self._size -= 1
        if self.isempty():
            self._head = None
            self._tail = None
        print(f'removed first element: {e}')
        return e

    def removeanywhere(self,position):
        if self.isempty():
            print('List is empty')
            return
        p = self._head
        index = 1
        while index < position:
            p = p._next
            index += 1
        e = p._next._element
        p._next = p._next._next
        self._size -= 1
        print(f'removed the element {e}')
        return e
        
    
    def removelast(self):
        if self.isempty():
            print('List is empty')
            return
        index = 1
        p = self._head
        
        while index < len(self) - 1:
            p = p._next
            index += 1
        self._tail = p
        p = p._next
        self._tail._next = self._head
        e = p._element
        self._size -= 1
        print(f'last element has been removed: {e}')
    
    def display(self):
        p = self._head
        i = 0
        while i < len(self):
            print(p._element,end='-->')
            p = p._next
            i += 1
        print()

    def search(self,key):
        p = self._head
        index = 0
        while index < len(self):
            if p._element == key:
                return index
            p = p._next
            index = index + 1
        return -1
```

# DOUBLY LINK LIST

#### 1. Doubly Linked List is like singly linked list but it also has the reference of the previous node

```python
class _Node:
    __slots__ = '_element', '_next', '_prev'
    
    def __init__(self, element, next, prev):
        self._element = element
        self._next = next
        self._prev = prev
        
class DoublyLinkedList:
    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0
    
    def __len__(self):
        return self._size
        
    def isempty(self):
        return self._size == 0
    
    def addlast(self, e):
        newest = _Node(e,None,None)
        
        if self.isempty():
            self._head = newest
            self._tail = newest
        else:
            self._tail._next = newest
            newest._prev = self._tail
            self._tail = newest
        self._size += 1
        
    def addfirst(self, e):
        newest = _Node(e,None,None)
        
        if self.isempty():
            self._tail = newest
            self._head = newest
        else:
            newest._next = self._head
            self._head._prev = newest
            self._head = newest
        self._size += 1
        
    def addanywhere(self, e, position):
        newest = _Node(e,None,None)
        p = self._head
        index = 1
        
        while index < position:
            p = p._next
            index += 1
            
        newest._next = p._next
        p._next._prev = newest
        p._next = newest
        newest._prev = p
        self._size += 1
        
    def removefirst(self):
        if self.isempty():
            print('The list is empty')
            return
        
        e = self._head._element
        self._head = self._head._next
        self._head._prev = None
        self._size -= 1
        
        if self.isempty():
            self._tail = None
        print(f'Deleted first element {e}')
        
    def removelast(self):
        if self.isempty():
            print('The List Is empty')
            return
        
        e = self._tail._element
        self._tail = self._tail._prev
        self._tail._next = None
        self._size -= 1
        
        if self.isempty():
            self._head = None
            self._tail = None
        print(e)
        
    def removeanywhere(self, position):
        p = self._head
        index = 1
        
        if self.isempty():
            print("List is empty")
            return
            
        while index < position:
            p = p._next
            index += 1
        
        e = p._next._element
        p._next = p._next._next
        p._next._prev = None
        self._size -= 1
        print(e)
        
    def display(self):
        p = self._head
        while p:
            print(p._element,end='-->')
            p = p._next
        print()
        
    def display_reverse(self):
        p = self._tail
        while p:
            print(p._element,end='-->')
            p = p._prev
        print()
```
